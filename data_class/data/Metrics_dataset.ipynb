```python
import pandas as pd
import numpy as np
import researchpy as rp

pd.set_option('display.max_columns', 500)
```

```python
import pickle

def save_model(file_name, model):
    with open(file_name, 'wb') as file:
        pickle.dump(model, file)
```

```python
df_feature_envy = pd.read_csv('./data_class.csv')
df_feature_envy.drop(columns=['method'], inplace=True)
df_feature_envy.head()
```

```python
df_metrics = pd.read_excel('./metrics/MLCQ_DC_ck_metrics_from_project.xlsx')
df_metrics['from_project'] = np.where(df_metrics['from_project'], 1, 0)
df_metrics.head()
```

```python
df_metrics.shape
```

```python
df_merged = pd.merge(df_feature_envy, df_metrics, how='inner')
df_merged.head()
```

```python
df_merged['label'] = np.where(df_merged.severity == 'none', 0, 1)
df_merged.head()
```

```python
rp.summary_cat(df_merged["severity"])
```

```python
df_merged.drop(columns=['file', 'class'], inplace=True)
df_merged.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(sparse=False)

df_encoded = pd.DataFrame (encoder.fit_transform(df_merged[['type']]))

df_encoded.columns = encoder.get_feature_names(['type'])

df_merged.drop(['type'] ,axis=1, inplace=True)

df_merged= pd.concat([df_merged, df_encoded ], axis=1)
df_merged.head()
```

```python
df_merged.to_csv('./embedded_datasets/metrics_dataset.csv', index=False)
```

```python
df_merged.to_pickle('./embedded_datasets/metrics_dataset.pkl')
```
